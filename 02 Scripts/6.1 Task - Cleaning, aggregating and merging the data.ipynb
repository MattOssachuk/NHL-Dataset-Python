### Checking for missing values

```python
# Checking df_game
df_game.isnull().sum()
```

No missing values in df_game

```python
# Checking df_skater_stats
df_skater_stats.isnull().sum()
```

df_skater_stats has 398,107 missing values in the hits, takeaways, giveaways and blocked columns. This needs to be fixed.

Since the hits column does not have many values exceeding 10, and most hang around 1-3 I think using the mean of 1.3 would be appropriate here for imputing.

```python
#Replacing the hits column with the average
df_skater_stats['hits'] = df_skater_stats['hits'].fillna(df_skater_stats['hits'].mean())
```

Since the takeaways column does not usually exceed 4 I am going to impute using the average

```python
#Replacing the takeaways column with the average
df_skater_stats['takeaways'] = df_skater_stats['takeaways'].fillna(df_skater_stats['takeaways'].mean())
```

Since there are a lot of 0s in the give away column I think the average of 0.4 is appropriate here.

```python
#Replacing the giveaways column with the average
df_skater_stats['giveaways'] = df_skater_stats['giveaways'].fillna(df_skater_stats['giveaways'].mean())
```

The average seems appropriate for the blocked column as well. Going to impute the average.

```python
#Replacing the blocked column with the average
df_skater_stats['blocked'] = df_skater_stats['blocked'].fillna(df_skater_stats['blocked'].mean())
```

```python
# Re-Checking df_skater_stats
df_skater_stats.isnull().sum()
```

Success!
